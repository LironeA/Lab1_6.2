```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_area(num_points=10000):
    x_min, x_max = -1.25, 1.25
    y_min, y_max = -1, 1.75

    x_points = np.random.uniform(x_min, x_max, num_points)
    y_points = np.random.uniform(y_min, y_max, num_points)

    inside_points = (x_points**2 + (y_points - np.sqrt(np.abs(x_points)))**2 <= 1)

    area_rectangle = (x_max - x_min) * (y_max - y_min)
    area_shape = area_rectangle * np.sum(inside_points) / num_points

    visualize_monte_carlo(x_points, y_points, inside_points, x_min, x_max, y_min, y_max, area_shape)
    
    return area_shape

def visualize_monte_carlo(x_points, y_points, inside_points, x_min, x_max, y_min, y_max, area_shape):
    plt.figure(figsize=(8, 8))

    plt.scatter(x_points[inside_points], y_points[inside_points], color='blue', s=0.5, label="Inside Figure")

    plt.scatter(x_points[~inside_points], y_points[~inside_points], color='red', s=0.5, label="Outside Figure")

    x_contour = np.linspace(x_min, x_max, 1000)
    y_contour_upper = np.sqrt(1 - x_contour**2) + np.sqrt(np.abs(x_contour))
    y_contour_lower = -np.sqrt(1 - x_contour**2) + np.sqrt(np.abs(x_contour))
    plt.plot(x_contour, y_contour_upper, color='black', linewidth=1)
    plt.plot(x_contour, y_contour_lower, color='black', linewidth=1)

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"Оцінена площа фігури методом Монте-Карло: {area_shape}")
    plt.legend()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

estimated_area = monte_carlo_area()
print(f"Оцінена площа фігури: {estimated_area}")
```
